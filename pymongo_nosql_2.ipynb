```python
import numpy as np
import pandas as pd
import pymongo
```

```python
# establish a connection with MongoDB
# connect to stock.apple
conn=pymongo.MongoClient('localhost',27017)
apple=pd.DataFrame(list(conn['stock'].apple.find()))
# apple=pd.DataFrame.from_records(conn['stock'].apple.find())
apple.head()
```

```python
from odo import odo
apple_odo=odo('mongodb://localhost/stock::apple',pd.DataFrame)
apple_odo.head()
```

```python
# make variable as index
apple.set_index('Date',inplace=True)
apple.sort_index(inplace=True)
apple.head()
```

```python
# choose the records with 'Close' great than 130
meta=conn['stock'].apple.find({'Close':{'$gt':130}})
apple_cond=pd.DataFrame.from_records(meta)
apple_cond.head()
```

```python
# choose the records with 'Close' great than 130 and 'Open' less than 140 
meta=conn['stock'].apple.find({'$and':[{'Close':{'$gt':130}},
                                      {'Open':{'$lt':140}}]})
apple_cond_2=pd.DataFrame.from_records(meta)
apple_cond_2.head()
```
